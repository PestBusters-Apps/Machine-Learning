```python
# Import the necessary libraries
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# Define the paths
model_path = "final_train.pt"
image_path = "Screenshot 2024-11-27 080014.png"

# Check if paths exist
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model path '{model_path}' does not exist.")
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Image path '{image_path}' does not exist.")

# Load a pretrained YOLOv8 model
model = YOLO(model_path)

# Predict on an image
results = model.predict(source=image_path, conf=0.25, save=False)  # Set save=False to avoid saving to disk

# Extract the annotated image from results
annotated_image = results[0].plot()

# Display the annotated image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Detection Results")
plt.show()
```
